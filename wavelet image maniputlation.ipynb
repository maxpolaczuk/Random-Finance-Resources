# Wavelet Decomposition of Donald Trump-kun :3, Kawaii-desu

```python
%matplotlib inline
```

```python
import numpy as np
import pywt as pw
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import color
from skimage import io
```

```python
img = color.rgb2gray(io.imread('barb.png'))
```

```python
imgplot = plt.imshow(img, cmap = "gray")
```

```python
# now we look at the images shape...
img
```

## Use a DWT on the image

```python
cA1, cD3, cD2, cD1 = pw.wavedec(img[:,:], 'db4', level = 3 )
```

## visualise the first transformation

```python
#plt.imshow(l1)
fig = plt.figure(1)

plt.subplot(221)
plt.imshow(cD1, cmap = "gray",aspect='auto')
plt.title("Detail 1")
plt.axis('off')

plt.subplot(222)
plt.imshow(cD2, cmap = "gray",aspect='auto')
plt.title("Detail 2")
plt.axis('off')

plt.subplot(223)
plt.imshow(cD3, cmap = "gray",aspect='auto')
plt.title("Detail 3")
plt.axis('off')

plt.subplot(224)
plt.imshow(cA1, cmap = "gray",aspect='auto')
plt.title("Smoothed Image")
plt.axis('off')
```
